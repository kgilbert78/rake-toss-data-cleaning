```python
import pandas as pd
import numpy as np
from pathlib import Path

years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
```

```python
column_names = {}
longest = 0

# get column names for 2008 to 2023 table_A and table_B CSVs
for year in years:
  if year > 2008:
    dfA = pd.read_csv(f'output/table_A_{year}.csv')
    a_list = list(dfA.columns)
    column_names[f'{year}-A'] = a_list
    if len(a_list) > longest:
      longest = len(a_list)
  dfB = pd.read_csv(f'output/table_B_{year}.csv')
  b_list = list(dfB.columns)
  column_names[f'{year}-B'] = b_list
  if len(b_list) > longest:
      longest = len(b_list)

# table_A_car_pond_2013.csv and table_B_car_pond_2013.csv will have the same column names as the lake dataframes for 2013

print('longest', longest)
print(column_names)
```

```python
for key in column_names:
  if len(column_names[key]) < longest:
    for i in range(longest - len(column_names[key])):
      column_names[key].append('Blank')
```

```python
# make them into a dataframe that can be exported as a CSV
col_name_df = pd.DataFrame.from_dict(column_names)
col_name_df
```

```python
# work towards desired results with smaller dataframes
test_years = [2000, 2001, 2002, 2003]
df2000 = pd.DataFrame({'colors': ['red', 'orange', 'yellow'], 'numbers': [21, 43, 32], 'languages': ['english', 'french', 'spanish']})
df2001 = pd.DataFrame({'colors': ['green', 'blue', 'purple'], 'numbers': [11, 22, 54], 'animals': ['dog', 'cat', 'rabbit']})
df2002 = pd.DataFrame({'colors': ['white', 'pink', 'teal'], 'coding_languages': ['javascript', 'python', 'sql']})
df2002 = pd.DataFrame({'colors': ['black', 'brown', 'tan'], 'platforms': ['mac', 'linux', 'windows'], 'frameworks': ['django', 'react', 'vue'], 'cities': ['syracuse', 'buffalo', 'concord'], 'numbers': [44, 33, 55]})
df2003 = pd.DataFrame({'colors': ['gray', 'coral', 'navy'], 'instruments': ['guitar', 'piano', 'drums']})
all_test_years = {2000: df2000, 2001: df2001, 2002: df2002, 2003: df2003}

for i, year in enumerate(test_years):
  # flip through anchor years
  year + i
  a_list = list(all_test_years[year].columns)
  for col_name in a_list:
    # year is the same as annees[i]
    years_to_search = [yr for yr in test_years if yr != year]
    # print(year, 'years_to_search', years_to_search)
    # make new annees list without the current year
    for search_year in years_to_search:
      # flip through all the non-anchor years
      if col_name not in list(all_test_years[search_year].columns):
        print(i, year, f'{col_name} not in {search_year}')
        all_test_years[search_year][col_name] = None
```
