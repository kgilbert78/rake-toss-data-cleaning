```python
import pandas as pd
import numpy as np
from pathlib import Path

years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
```

```python
column_names = {}
longest = 0

# get column names for 2008 to 2022 table_A and table_B CSVs
for year in years:
  if year > 2008:
    dfA = pd.read_csv(f'output/table_A_{year}.csv')
    a_list = list(dfA.columns)
    column_names[f'{year}-A'] = a_list
    if len(a_list) > longest:
      longest = len(a_list)
  dfB = pd.read_csv(f'output/table_B_{year}.csv')
  b_list = list(dfB.columns)
  column_names[f'{year}-B'] = b_list
  if len(b_list) > longest:
      longest = len(b_list)

# table_A_car_pond_2013.csv and table_B_car_pond_2013.csv will have the same column names as the lake dataframes for 2013

print('longest', longest)
print(column_names)
```

```python
for key in column_names:
  if len(column_names[key]) < longest:
    for i in range(longest - len(column_names[key])):
      column_names[key].append('Blank')
```

```python
# make them into a dataframe that can be exported as a CSV
col_name_df = pd.DataFrame.from_dict(column_names)
col_name_df
```

```python
col_name_df.to_csv('output/ALL_COLUMN_NAMES.csv')
```
