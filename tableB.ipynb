Grab milfoil column from each year and make a Dataframe of it *(while waiting for feedback on latin name matching)*

```python
years = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
# 17-21 also need title line repeats
milfoil_dfs = {}

for year in years:
  current_df = all_dfs[year]
  current_df.dropna(axis = 0, how = 'all', inplace = True)
  
  if year > 2016:
    # remove rows with column names repeated (for scrolling in Excel)
    current_df.drop(current_df[current_df['Latitude'] == ' Latitude'].index, inplace = True)
  
  for index, row in current_df.iterrows():
    # print(year, row['Sample_Point_(SP)'], type(row['Sample_Point_(SP)']))
    new_sample_point = row['Sample_Point_(SP)'].replace('C', '')
    current_df['Sample_Point_(SP)'][index] = new_sample_point

  milfoil_dfs[year] = current_df.filter(items=['Sample_Point_(SP)', 'Latitude', 'Longitude', 'Myriophyllum_spicatum'])
  
  current_df.to_csv(f'output/table_B_{year}.csv')
  milfoil_dfs[year].to_csv(f'output/milfoil/table_B_{year}.csv')
  
```
