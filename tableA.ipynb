```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
df2009 = pd.read_excel("input/rakeToss2009-A.xls")
df2009
```

```python
for col in df2009.columns:
    df2009.rename(columns={col:col.strip().replace(" ","_")},inplace=True)
df2009.columns
```

```python
rake_toss_2_lines = np.where(df2009['Rake_toss_#'] == 2)
rake_toss_2_lines
```

```python
for line in rake_toss_2_lines:
    # copy each point's sample point number, latitude and longitude to the line below where they are blank
    df2009['Sample_Point_(SP)'][line] = df2009['Sample_Point_(SP)'][line-1]
    df2009['Latitude'][line] = df2009['Latitude'][line-1]
    df2009['Longitude'][line] = df2009['Longitude'][line-1]
    df2009['Depth_(ft)_August_2008_-_Allied_Biological'] = df2009['Depth_(ft)_August_2008_-_Allied_Biological'][line-1] # not working
    df2009['Depth_(ft)_Aug-Sept._2009_-_Racine_Johnson'] = df2009['Depth_(ft)_Aug-Sept._2009_-_Racine_Johnson'][line-1] # not working
    df2009['Depth_(m)_Aug-Sept._2009_-_Racine_Johnson'] = df2009['Depth_(m)_Aug-Sept._2009_-_Racine_Johnson'][line-1] # not working
df2009.head()
```

```python
df2009.dropna(axis = 0, how = 'all', inplace = True)
df2009.head()
```

```python
# remove rows with column names repeated (for scrolling in Excel)
df2009.drop(df2009[df2009['Latitude'] == ' Latitude'].index, inplace = True)
```

```python
df2009.columns
```

```python
df2009 = df2009.rename(columns={'Ranunculus_sp.': 'Ranunculus_trichophyllus'})
df2009 = df2009.rename(columns={'Hypericum_ellipticum)': 'Hypericum_ellipticum'})
df2009 = df2009.rename(columns={'Nitellopsis_obstusa)': 'Nitellopsis_obtusa'})
df2009.columns
```

```python
df2009.iloc[[27,28,29]]
```

```python
# df2009.loc[(df2009['Sample_Point_(SP)'].isna()) & (df2009['Total_%'] == 0)]
df2009.loc[df2009['Sample_Point_(SP)'].isna()]['Total_%']
```

```python
df2009.drop(df2009.loc[df2009['Sample_Point_(SP)'].isna()].index, inplace=True)
```

```python
df2009.iloc[[27,28,29]]['Total_%']
```

```python
for index, row in df2009.iterrows():
  # print(index, row['Sample_Point_(SP)'], type(row['Sample_Point_(SP)'])) # one row was NaN except Total_%
  new_sample_point = row['Sample_Point_(SP)'].replace('C', '')
  df2009['Sample_Point_(SP)'][index] = new_sample_point
df2009
```

```python
df2009.to_csv('output/table_A_2009.csv')
```
